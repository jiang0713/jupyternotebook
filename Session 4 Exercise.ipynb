```python
#Step 2
temp_start = input('Enter the starting temperature?')
temp_end = input('Enter the ending temperature?')
```

```python
#Step 3
wind_start = input('Enter the starting wind speed?')
wind_end = input('Enter the ending wind speed?')
```

```python
#Step 4
step = input('Enter the step increment for the temperature?')
```

```python
#Step 5

def CalculateWindChill(temp, wind):
    chill = round((13.12 + (0.6215 * temp) - (11.37 * (wind **0.16)) + (0.3965 * temp * (wind ** 0.16))), 3)
    return chill
```

```python
#Step 6
temp_start = int(temp_start)
temp_end = int(temp_end)
wind_start = int(wind_start)
wind_end = int(wind_end)
step = int(step)
temp_now = temp_start
while temp_now <= temp_end:
    print('temperature is '+ str(temp_now))
    wind_now = wind_start
    while wind_now <= wind_end:
        chill_now = CalculateWindChill(temp = temp_now, wind = wind_now)
        print('    wind is ' + str(wind_now) + ' calculated wind chill is: ' + str(chill_now))
        wind_now += step
    temp_now += step
```
